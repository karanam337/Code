```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
dataset
```

```python
basket = list()
for transaction in dataset:
    for item in transaction: 
        basket.append(item)
print(basket)
print(set(basket))
print(len(set(basket)))
```

```python
unique_items = set([item for transaction in dataset for item in transaction])
unique_items_len = len(unique_items)
dataset_len = len(dataset)

print(unique_items)
print('unique_items_len: ', unique_items_len)
print('dataset_len: ', dataset_len)
```

```python
import numpy as np

matrix = np.zeros((dataset_len, unique_items_len))
print(matrix)
```

```python
for row, transaction in enumerate(dataset): 
    #print(row, transaction)
    for item in transaction: 
        index = list(unique_items).index(item)
        matrix[row, index] = 1
    #print(matrix[row])
print(matrix)
```

```python
import pandas as pd

basket_matrix = pd.DataFrame(matrix, columns = list(unique_items))
basket_matrix
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori 

one_hot = TransactionEncoder()
one_hot_matrix = one_hot.fit(dataset).transform(dataset)
print(one_hot_matrix, type(one_hot_matrix))

print('\n')
print(one_hot.columns_)

df = pd.DataFrame(one_hot_matrix, columns = one_hot.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support = 0.6, use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets_sorted = frequent_itemsets.sort_values('support', ascending = False)
frequent_itemsets_sorted
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric = 'lift', min_threshold = 1.2)
rules
```
