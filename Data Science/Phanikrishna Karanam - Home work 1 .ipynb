__Summary__: Apply the apriori algorithm and association rules to groceries.xlsx dataset using MLXtend and obtain below: 
 - Top 10 itemsets
 - Top 10 rules based on Lift
 
Note: groceries.xlsx excel to be placed in the same location from where python notebook is being executed from

 - Import the Pandas library and read the input excel file into a dataframe

```python
import pandas as pd

dataset_df = pd.read_excel('groceries.xlsx', header=None)
dataset_df.head()
```

 - Convert the dataframe to list display the first 5 elements of the list

```python
dataset_list = dataset_df.values.tolist()
print(dataset_list[:5])                      
```

 - Loop thru the list to remove Null values (represented by nan) and display first 5 elements of the list

```python
marketbasket_db = []
for itemset in dataset_list:
    templist = []
    for item in itemset: 
        if str(item) != 'nan': 
            templist.append(item)
    marketbasket_db.append(templist)

print(marketbasket_db[:5])
```

 - Import TransactionEncoder from mlxtend's preprocessing library
 - Encode the transactions in database using Transaction Encoder, Fit & Transform 
 - Convert the encoded array into a Dataframe and display the top few records

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_arr = te.fit(marketbasket_db).transform(marketbasket_db)
df = pd.DataFrame(te_arr, columns = te.columns_)
df.head()
```

 - Import apriori from mlxtend's frequent patterns library
 - Pass the encoded dataframe to apriori and set the minimum support value as desired 
 - Display the frequent itemsets as returned by algorithm 

```python
from mlxtend.frequent_patterns import apriori

freq_itemsets = apriori(df, min_support = 0.045, use_colnames = True)
freq_itemsets
```

 - Sort in descending order of Support and display top 10 itemsets

```python
freq_itemsets_sorted = freq_itemsets.sort_values('support', ascending = False)
print('Frequent Itemsets sorted by highest Support:')
freq_itemsets_sorted.head(10)
```

 - Import association rules from mlxtend's frequent patterns library
 - Pass the previously generated frequent itemsets and set the minimum threshold value for Lift metric as desired
 - Display the rules generated by algorithm

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(freq_itemsets, metric = 'lift', min_threshold = 1.2)
rules
```

 - Sort in descending order of Lift and display top 10 rules

```python
rules_sorted = rules.sort_values('lift', ascending = False)
print('Top 10 rules sorted by highest Lift:')
rules_sorted.head(10)
```
