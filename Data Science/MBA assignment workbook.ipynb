```python
import pandas as pd

dataset_df = pd.read_excel('groceries.xlsx', header=None)
dataset_df.head(5)
```

```python
dataset_list = dataset_df.values.tolist()
#print(dataset_list[:10])
#print(len(dataset_list))

print(dataset_list[:5])
```

```python
dataset = [item for transaction in dataset_list for item in transaction if str(item) != 'nan']
dataset[:10] 
```

```python
for transaction in dataset_list: 
    print(transaction)
    temp = transaction[~pd.isnull(transaction)]
    print(temp)
```

```python
dataset = []
for transaction in dataset_list:
    templist = []
    for item in transaction: 
        if str(item) != 'nan': 
            templist.append(item)
    dataset.append(templist)
print(dataset[:10])
```

```python
dataset_arr = dataset_df.to_numpy()
print(dataset_arr)
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_arr = te.fit(dataset).transform(dataset)
print(te_arr)
df = pd.DataFrame(te_arr, columns = te.columns_)
df.head()
```

```python
from mlxtend.frequent_patterns import apriori

freq_itemsets = apriori(df, min_support = 0.045, use_colnames = True)
freq_itemsets
```

```python
freq_itemsets_sorted = freq_itemsets.sort_values('support', ascending = False)
print('Frequent Itemsets sorted by highest Support: \n')
freq_itemsets_sorted.head(10)
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(freq_itemsets, metric = 'lift', min_threshold = 1.2)
rules
```

```python
rules_sorted = rules.sort_values('lift', ascending = False)
print('Top 10 rules sorted by highest Lift: \n')
rules_sorted.head(10)
```
