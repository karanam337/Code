```
data = spark.read.csv("databricks-datasets/flights/departuredelays.csv",
                     header = 'true')
data.show(5)
```

```
airports_data = spark.read.csv("databricks-datasets/flights/airport-codes-na.txt",
                     header = 'true', inferSchema = 'true', sep = '\t')
airports_data.show(3, truncate = False)
```

```
#table for flights
data.createOrReplaceTempView("flights")
#table for airports
airports_data.createOrReplaceTempView("airports")
```

```
#Sum of flight delays by city and origin code for the state of California
spark.sql("""
select a.city,
f.origin,
sum(f.delay) as Delays
from flights f
join airports a
on a.IATA = f.origin
where a.state = 'CA'
group by a.city, f.origin
order by Delays desc limit 5""").show()
```

```
%sql
select a.city,
f.origin,
sum(f.delay) as Delays
from flights f
join airports a
on a.IATA = f.origin
where a.state = 'CA'
group by a.city, f.origin
order by Delays desc
```
