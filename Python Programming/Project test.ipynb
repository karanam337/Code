```python
import urllib.request, urllib.parse, urllib.error
import re 
import sqlite3

cnn_fhand = urllib.request.urlopen('https://money.cnn.com/data/hotstocks/')
cnn_html_str = cnn_fhand.read().decode()
cnn_html_str = cnn_html_str.replace('\n','').replace('\r','').replace('\t','') 
active_list = re.findall('<h3>Most Actives</h3>.*?</table>', cnn_html_str)
active_tickers = re.findall('class="wsod_symbol">([^ <]*)', active_list[0])
print('The active tickers on CNN site are: ', active_tickers)

yf_url  = 'https://finance.yahoo.com/quote/{tick}?p={tick}&.tsrc=fin-srch-v1'

main_list = []
file_write_list = []
for ticker in active_tickers:
    yf_fhand = urllib.request.urlopen(yf_url.format(tick = ticker))
    yf_html_str = yf_fhand.read().decode()
    yf_html_str = yf_html_str.replace('\n','').replace('\r','').replace('\t','')
#    tickdetails_list = re.findall('PREV_CLOSE-value.*?Avg', yf_html_str)
    temp_list = []
    temp_list.append(ticker)
    
    yf_search = 'OPEN-value'
    temp = re.findall(yf_search + '.*?([^ >]*)</span></td></tr>',yf_html_str)
    temp_list.append(float(temp[0]))
    
    yf_search = 'PREV_CLOSE-value'
    temp = re.findall(yf_search + '.*?([^ >]*)</span></td></tr>',yf_html_str)
    temp_list.append(float(temp[0]))

    yf_search = 'TD_VOLUME-value'
    temp = re.findall(yf_search + '.*?([^ >]*)</span></td></tr>',yf_html_str)
    temp_list.append(int(temp[0].replace(',','')))
    
    temp_list.append('\n')
    
    print('temp_list :', temp_list)
    
    main_list.append(temp_list)
    file_write_list.append(((',').join(map(str,temp_list))))

print('main_list: ', main_list)
print('file_write_list: ', file_write_list)

fhand = open('stocks.txt','w')
fhand.writelines(file_write_list)
fhand.close()

for sub_list in main_list:
    sub_list.pop()

print('main_list: ', main_list)
    
conn = sqlite3.connect('StocksDatabase.sqlite')
cur = conn.cursor()
column_names = '(Ticker TEXT, OpenPrice REAL, PrevClose REAL, Volume INTEGER)'
cur.execute('DROP TABLE IF EXISTS StocksTable')
cur.execute('CREATE TABLE StocksTable' + column_names) 
cur.executemany('INSERT INTO StocksTable VALUES (?,?,?,?);', main_list) 
conn.commit()
cur.close()
conn.close()
```

```python
import urllib.request, urllib.parse, urllib.error
import re 
import sqlite3
import time

start_time = time.time()
print('Program execution started') 
website_up = True

try: 
    cnn_fhand = urllib.request.urlopen('https://money.cnn.com/data/hotstocks/')
except:  
    print('CNN website is currently down. Please try again') 
else: 
    cnn_html_str = cnn_fhand.read().decode()
    active_list = re.findall('<h3>Most Actives</h3>.*?</table>', cnn_html_str, re.DOTALL)
    active_tickers = re.findall('class="wsod_symbol">([^ <]*)', active_list[0])

    yf_url  = 'https://finance.yahoo.com/quote/{tick}?p={tick}&.tsrc=fin-srch-v1'
    main_list = []
    file_write_list = []

    for ticker in active_tickers:  
        yf_html_list = []
        temp_list = []

        try: 
            yf_fhand = urllib.request.urlopen(yf_url.format(tick = ticker))
        except: 
            print('Yahoo Finance website is currently down. Please try again')
            website_up = False
            break 
        else: 
            yf_html_str = yf_fhand.read().decode()
            yf_html_list = re.findall('PREV_CLOSE-value.*?Avg', yf_html_str, re.DOTALL)

            temp_list.append(ticker)

            temp = re.findall('OPEN-value.*?([^ >]*)</span></td></tr>',yf_html_list[0], re.DOTALL)
            temp_list.append(float(temp[0]))

            temp = re.findall('PREV_CLOSE-value.*?([^ >]*)</span></td></tr>',yf_html_list[0], re.DOTALL)
            temp_list.append(float(temp[0]))

            temp = re.findall('TD_VOLUME-value.*?([^ >]*)</span></td></tr>',yf_html_list[0], re.DOTALL)
            temp_list.append(int(temp[0].replace(',','')))

            main_list.append(temp_list)
            file_write_list.append(((',').join(map(str,temp_list))) + '\n')
             
    if website_up:
        fhand = open('stocks.txt','w')
        fhand.writelines(file_write_list)
        fhand.close()

        conn = sqlite3.connect('StocksDatabase.sqlite')
        cur = conn.cursor()
        
        column_names = '(Ticker TEXT, OpenPrice REAL, PrevClose REAL, Volume INTEGER)'
        
        cur.execute('DROP TABLE IF EXISTS StocksTable')
        cur.execute('CREATE TABLE StocksTable' + column_names) 
        
        cur.executemany('INSERT INTO StocksTable VALUES (?,?,?,?);', main_list) 
        cur.close()
        
        conn.commit()
        conn.close()
        
        print('Elapsed time:', time.strftime('%H:%M:%S', time.gmtime(time.time() - start_time)))
        
print('Program execution ended')
```
