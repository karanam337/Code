##### __Expected Results__:
Average, maximum, minimum and median will be displayed for the entered 3 or 4 integers
    - For Median value, if number of integers entered are:
        - Even -> then average of the two middle values will be displayed as median 
        - Odd -> then middle integer will be displayed as median 
    - Average and Median value would be rounded to the closest integer depending on fractional part
      - If fractional value is >= 0.5, then round up (vice-versa for negative integer) 
      - If fractional value is < 0.5, then round down (vice-versa for negative integer) 
         
##### __Algorithm__:
` 1. Accept input choice from user whether they want to key in 3 or 4 integers

     3.1 Typecast the function parameter variable to integer, to get numeric part
     3.2 Subtract the parameter variable with numeric part obtained from above step, to get decimal part alone
     Note: as integer can have both postive and negative values, should consider both cases while rounding up
     3.3 For positive integer: If decimal part
         3.3.1 is >= 0.5, round up to next integer by adding 1 to integer parameter variable
         3.3.2 is < 0.5 or zero, then only numeric part (INT) of parameter variable will be considered
     3.4 For negative integer: If decimal part
         3.4.1 is <= -0.5, round down to next integer by subtracting 1 from integer parameter variable  
         3.4.2 is > -0.5 or zero, then only numeric part (INT) of parameter variable will be considered
     3.5 Return the updated parameter value from the function `    
` 4. To compute Average:
     4.1 Calculate length of the stored list created in step 2.1 (will be 3 or 4, based on input choice)
     4.2 Calculate the sum of the list
     4.3 Divide sum of list by length of list for average
     4.4 Call round up function defined in step 3
     4.5 Display the returned value as average of the 3 or 4 numbers (based on input choice) `  
` 5. For Maximum: use the inbuilt max function and display the result `  
` 6. For Minimum: use the inbuilt min function and display the result `  
` 7. To compute Median, first sort the list in ascending order using built in sort function
     7.1 Check if length of the list is divisible by 2 (i.e input choice is 4)
         7.1.1 If remainder is zero, then determine average of middle 2 numbers (position 1 and 2 in list)
               7.1.1.1 Call round up function defined in step 3 to round up median
         7.1.2 If remainder is not zero (i.e. input choice is 3), then middle number (position 1 in list) is median
     7.2 Display the median calculated `  

```python
# Program for Question 2

# Prompt user to enter their choice of either 3 or 4 integers
# Handles all invalid inputs like string, float, negative numbers
# Operates in a conversational mode until valid input is entered 

while_flag = 0
while (while_flag == 0):
    try:
        input_choice = input("Enter set of 3 or 4 integers: ")
        int_input_choice = int(input_choice)  
    except ValueError:
        print("Invalid integer. Valid inputs are 3 or 4 only")
    else:
        if ((int_input_choice == 3) or (int_input_choice == 4)):
            while_flag = 1
        else:
            print("Invalid input. Valid inputs are 3 or 4 only")

# Depending on input choice, prompt for either 3 or 4 input integers (could be positive or negative) 
# Accepts one input integer at a time and stores in a list variable 
# Handles all invalid inputs like string and float 
# Operates in a conversational mode until valid input is entered
# Execution will resume from same point after a valid input is entered. User need not execute the program from start again
            
while_flag = 1       # to determine when to exit the while loop
num_counter = 1      # to determine which input integer (1, 2, 3 or 4) is being inputed 
number_list = list() # store entered numbers in this list 

while (while_flag <= int_input_choice):
    try: 
        tempnum = input("Enter number {}: ".format(num_counter))      
        int_tempnum = int(tempnum)  
    except ValueError:
        print("Invalid integer. Please re-enter") 
    else:
        number_list.append(int_tempnum) 
        while_flag +=1
        num_counter +=1

# Function to round the float number to nearest integer, based on decimal value 
# Considers both positive and negative integers and rounds it accordingly 
# For positive integer: If fractional value >= 0.5, then round up, else round down to nearest integer
# For negative integer: If fractional value <= -0.5, then round up, else round down to nearest integer
# If number is an integer, then no rounding will be done
# Float value being passed to function and Integer value being returned from function 

def round_int(parm_number):  
    
    int_parm_number = int(parm_number)                              # to get numeric part only 
    fraction_parm_number = parm_number - int_parm_number            # to determine fractional part 
    
    if (int_parm_number >= 0) and (fraction_parm_number >= 0.5):    # for positive integers round up
        int_parm_number +=1
    elif (int_parm_number <= 0) and (fraction_parm_number <= -0.5): # for negative integers round down
        int_parm_number -=1
        
    return int_parm_number

# Compute Average after determinig length and sum of the list 
# Call round_int function which is defined above to round the value to nearest integer

list_len = len(number_list) 
avg_number = sum(number_list) / list_len 
int_avg_number = round_int(avg_number)
num_1 = number_list[0]
num_2 = number_list[1]
num_3 = number_list[2]

if (int_input_choice == 3):
    print("The average of {}, {} and {} is:".format(num_1, num_2, num_3), int_avg_number) 
else:
    num_4 = number_list[3]
    print("The average of {}, {}, {} and {} is:".format(num_1, num_2, num_3, num_4), int_avg_number) 

# use inbuilt max function to determine the highest number in the list as entered by user 
max_number = max(number_list)
print("The maximum of the {} numbers is:".format(int_input_choice), max_number)

# use inbuilt min function to determine the highest number in the list as entered by user 
min_number = min(number_list)
print("The minimum of the {} numbers is:".format(int_input_choice), min_number)

# Compute Median of the list by first sorting the list in ascending order
# If the length of the list is divisible by 2 (i.e even numbers keyed-in or input choice is 4), 
#  - Calculated average of both middle numbers will be the median
#  - Call round_int function to round the number to nearest integer
# If length of list is not divisible by 2 (i.e odd numbers keyed-in or input choice is 3), then middle number is median

number_list_temp = number_list     
number_list_temp.sort()
    
if list_len % 2 == 0:                                  # for even number of input choice
    median_1 = number_list_temp[list_len // 2]         # gets 3rd number from list. Note that list starting position is 0
    median_2 = number_list_temp[(list_len // 2) - 1]   # gets 2nd number from list 
    temp_median = (median_1 + median_2) / 2            # average of 2 middle numbers 
    int_median = round_int(temp_median)                # call round_int function to round to nearest integer 
else: 
    int_median = number_list_temp[list_len // 2]       # for odd number of input choice
print("The median of the {} numbers is:".format(int_input_choice), int_median)

# End of Program for Question 2
```

#### __Question 3__:  
__Summary__: Accept a positive integer from user and determine if its a prime, composite or neither prime / composite number

##### __Execution Instructions__:
1. Run the cell below   
2. Enter a positive integer  
3. To rerun the test, run the code cell again and follow the prompts

##### __Expected Results__:
 - Depending on the positive integer entered, the program will display if its a prime, composite or neither prime nor composite 
 - If an invalid input is entered, an error message will be displayed 
         
##### __Algorithm__:
` 1. Accept input from user and check whether it is numeric. Isnumeric check will test if its positive integer only 
     1.1. Convert the input string variable to integer format 
     1.2. If input is 0 or 1, then input is neither prime nor composite number
     1.3. If input is > 1, then starting from 2, perform below in a loop until input-1 as range 
         1.3.1. Check if input is perfect divisible (remainder zero) of any number starting from 2 until the range 
                1.3.1.1 If divisible, then input is a composite number. Break the program execution
         1.3.2. If number is not divisible by any other number until end of range (which is input -1), then number is prime `  
` 2. If entered input is not numeric, then display invalid input entered. Only positive integers are accepted `  

```python
# Program for question 3

input_number = input("Enter positive integer: ")             # Accept input from user 

if input_number.isnumeric():                                 # Check if entered input is a positive integer only
    int_input_number = int(input_number)                     # typecast input string to integer
    if ((int_input_number == 0) or (int_input_number == 1)): # 0 and 1 are neither prime nor composite
        print("It is neither prime nor composite")
    elif int_input_number > 1: 
        for i in range(2, int_input_number):      # check if input is divisible by any number other than itself and 1 or 0
                                                  # range function by default takes input-1 as end            
            if (int_input_number % i) == 0:       # if divisible with remainder as zero, then its a composite number
                print("It is a composite")  
                break                             # exit the program execution 
        else: 
            print("It is a prime")                # if not divisible by any other number, then its a prime number
else:
    print("You did not enter a positive integer") # when invalid inputs like strings or negative integers are entered

# End of program for question 3
```

     - Quotient value as per division operation

##### __Algorithm__:
` 1. Define SmartDivision function that accepts two parameters (dividend and divisor)
     1.1. Calculate the quotient and also determine the remainder
     1.2. If remainder=0, then typecast the quotient variable to integer (as its a whole number)
     1.3. Else quotient will remain as float variable
     1.4. Return the function result `  
` 2. Accept inputs from user and convert them to integers 
     2.1. Handle input errors when strings are entered or when 2nd input (i.e. divisor) is zero `  
` 3. If inputs are valid, then call SmartDivision function and pass both integers as function input parameters

Note: within same cell smart division function is defined. Then below, program calling the defined function is written

```python
# Program for question 4

# SmartDivision is a function which will accept two integer parameters and perform division operation
# Function will return quotient value of division operation. If the number is perfect divisible (with remainder as zero): 
#  - Then quotient will be returned as integer type
#  - else quotient will be returned as float type 
# Below function will be called within the same cell 

def SmartDivision(num, den):
    quotient = num / den
    remainder = num % den
    if (remainder == 0):
        quotient = int(quotient) 
    return quotient

# Program that will call the above defined function:
# Accept both inputs and typecast to integer (handle for both positive and negative values)
# If either of inputs are non-integers, then handle input value exceptions (strings and float inputs are invalid)
# Check if the second input (divisor) is not=0. This is to avoid Zero Division Error 
# If divisor is not=0, then call the SmartDivision function by passing the input integer parameters
# The function will return the Quotient of division operation and variable type will be either Integer or Float
# Display results of the function call 

try: 
    inp1 = input("Enter an integer (Dividend): ")                   
    int_inp1 = int(inp1)                                               
    inp2 = input("Enter a non-zero integer (Divisor): ")         
    int_inp2 = int(inp2)                                               
except ValueError: 
    print("Invalid input value entered. Only integers are accepted")
else:     
    if int_inp2 == 0:
        print("Divisor cannot be zero. Only non-zero integers are accepted") 
    else: 
        result = SmartDivision(int_inp1, int_inp2)
        print("Division result of {}/{} = {} and return type is: {}".format(int_inp1,int_inp2,result,type(result)))

# End of program for question 4
```

##### __Algorithm__:
` 1. Accept 3 sets of co-ordinates ((x1,y1), (x2,y2) and (x,y))

                 2.1.3.2. If N/n -> display 'Good bye!' and exit program
                 2.1.3.3. If invalid input entered, prompt user to reenter until valid input is entered `  

```python
# Program for question 5

# Initialize variables 
axis_point = 'x'      # of defining x/y value dynamically during input prompt
num_counter = 1       # for displaying first or second set of points dynamically during input prompt 
input_counter = 1     # to know which input is being keyed in 
co_ordinates = list() # to store all the 6 input values

# Iteratively prompt user to enter 6 values (2 opposite co-ordinates of rectangle and 3rd point)
# When invalid values entered, prompt for reentry and resume execution from point of error 

while True: 
    try: 
        input_point = float(input("Enter {}{}: ".format(axis_point, num_counter)))  # accept input points
    except ValueError: 
        print("Invalid input. Enter only Float or Integer value") # handle invalid inputs like string 
    else: 
        co_ordinates.append(input_point)                          # store the input value in a list
        input_counter +=1                                         # increment input counter
        if (input_counter % 2 == 0):                              # for 2nd, 4th and 6th inputs, prompt for y-axis value
            axis_point = 'y'                
        else:  
            axis_point = 'x'                                      # for 1st, 3rd and 5th inputs, prompt for x-axis value
        
        if (3 <= input_counter <= 4):                             # for 3rd and 4th inputs, prompt for 2nd co-ordinate point
            num_counter = 2
        elif (4 <= input_counter <=6): 
            num_counter = ' '                                     # for 5th and 6th inputs, prompts only x,y value 
        elif input_counter > 6:                                   # once all 6 valid inputs are entered, enter inside 
            x1 = co_ordinates[0] 
            y1 = co_ordinates[1] 
            x2 = co_ordinates[2] 
            y2 = co_ordinates[3] 
            x  = co_ordinates[4] 
            y  = co_ordinates[5] 
            print("Entered co-ordinates: (x1 = {}, y1 = {}), (x2 = {}, y2 = {}), (x = {}, y = {})".format(x1,y1,x2,y2,x,y)) 
        
            if (((x1 <= x <= x2) or (x2 <= x <= x1)) and ((y1 <= y <= y2) or (y2 <= y <= y1))):
                print("True")                                           # if 3rd point within or on boundary of rectangle 
            else: 
                print("False")                                          # if 3rd point outside of rectangle 
            
            while_flag = 1
            while (while_flag == 1): 
                run_continue = input("Do you want to continue (Y/N): ") # prompt user to check if they want to continue
                if (run_continue == 'Y' or run_continue == 'y'):        # also caring for case insensitive input option
                    while_flag = 2                                      # exit criteria for while loop when Y or y input
                elif (run_continue == 'N' or run_continue == 'n'):      # exit criteria for while loop when N or n input
                    while_flag = 3
                else: 
                    print("Invalid value. Enter Y / N: ")               # loop until valid input is entered
                    continue 
              
            if (while_flag == 2):                                       # if continuing, then initialize all variables
                axis_point = 'x'                                        # program will continue from first while loop
                num_counter = 1
                input_counter = 1
            else: 
                print("Good bye!")                                      # if not continuing, then end program
                break

# End of program for question 5
```
