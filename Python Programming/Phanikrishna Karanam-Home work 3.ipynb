```python
# Start of Program 2 

# Function to build the dictionary from input file 
def build_dict(f_handle):                                        # Accept the file handle parameter
    f_comma_count = 0                                            # Initialize comma counter   
    f_dict={}                                                    # Create empty dict   
    for line in f_handle:                                        # loop thru the file. Reads one line (upto \n) at a time 
        if line.isspace():                                       # If the line is empty or spaces, then skip processing 
            continue
        else: 
            f_dict_temp={}                                       # Create empty temporary dict   
            temp_line = line.strip()                             # Strip leading and training white spaces
            f_comma_count = temp_line.count(',')                 # Count the number of commas in the line   
            if f_comma_count > 1:                                # If atleast 2 commas are present, then execute below
                line_list = temp_line.split(',')                 # Split the line into a list delimited by comma
                for i in range(1,len(line_list),2):              # Loop thru list starting from 2nd element, increment by 2
                    try: 
                        review_point = float(line_list[i+1])     # Convert the score to a float value
                    except: 
                        f_comma_count = -2                       # If score has invalid value, then set counter to -2 
                        break                                    # break from loop due to invalid value in score   
                    else:                                        # If format of file record is good, then execute below
                        f_dict_temp[line_list[i]] = review_point # Store the movie name and its score in temp dict 
                f_dict[line_list[0]] = f_dict_temp               # Key is reviewer name and value is temp dict
                                                                 # Temp dict contains movies and review scores
            else: 
                f_comma_count = -1                               # When input record doesnt contain comma delimiter
                break                                            # Break from the main for loop 
        if f_comma_count == -2:                                  # When input record contains invalid value in movie score 
            break                                                # Break from the main for loop 
    f_handle.close()                                             # Close the input file     
    return(f_dict,f_comma_count)                                 # Return the dict and comma counter  

# Function to prompt user to enter their choice and loop until valid choice is entered
def prompt_choice():                                              
    f_user_choice = input('What do you want to do? Input 1 for similarity between two reviewers, or Input 2 for similarity between one reviewer and all others in the database or 3 to quit: ')
    while True: 
        if (f_user_choice == '1' or f_user_choice == '2' or f_user_choice == '3'):  
            break                                                # End loop if valid choice is entered   
        else:                                                    # If invalid choice entered, prompt again to enter
            f_user_choice = input('Invalid input choice. Please enter 1 or 2 or 3: ')
    return (f_user_choice)                                       # Return the entered user choice    

# Function to perform case insensitive dictionary lookup
def case_insensitive(dict_inp , rev):                                  
    for i in dict_inp:                                           # Loop thru the dictionary 
        if rev.lower() == i.lower():                             # Check if user entered input matches to key in dict   
            return(True)                                         # If reviewer is present, then True as functionr response 

# Function to build a sub dictionary, where movies and score are returned as separate dictionary for matching reviewer 
def build_subdict(dict , f_rev):                                   
    for keys, values in dict.items():                            # Loop thru the main dictionary  
        if f_rev.lower() == keys.lower():                        # For matching reviewer name, retrieve the values
            f_keys_temp = keys                                   # Return the reviewer name as in dictionary for display 
            f_dict_rev = values                                  # Store as separate dictionary 
    return(f_dict_rev , f_keys_temp)                             # Return sub dict and reviewer name as in dict

# Function when option 1 is selected 
def option_1(f_dict):                                            # Accept the dict as parameter 
    sim_score_temp = 0                                           # Initialize the score variables  
    sim_score = 0   
    rev1 = input('Provide Reviewer1 name: ')                     # Accept Reviewer 1  
    rev2 = input('Provide Reviewer2 name: ')                     # Accept Reviewer 2
    if rev1.isspace() or len(rev1) == 0:                         # Check to see if valid input is entered
        print('Reviewer1 name not entered')                        
    elif not (case_insensitive(f_dict, rev1)):                   # Check if Reviewer1 is present in database
        print('Reviewer1 is not present in database')                        
    elif rev2.isspace() or len(rev2) == 0:                       # Check to see if valid input is entered    
        print('Reviewer2 name not entered') 
    elif not (case_insensitive(f_dict, rev2)):                   # Check if Reviewer2 is present in database
        print('Reviewer2 is not present in database') 
    else:    
        dict_rev1 , keys_temp1 = build_subdict(f_dict , rev1)    # Call to build a sub dict by passing reviewer 1 and dict 
        dict_rev2 , keys_temp2 = build_subdict(f_dict , rev2)    # Call to build a sub dict by passing reviewer 1 and dict 
        for keys, values in dict_rev1.items():                   # Loop thru the sub dict of reviewer 1  
            if keys in dict_rev2:                                # If movie name is present in reviewer 2 sub dict, then    
                sim_score_temp = (values - dict_rev2[keys]) ** 2 # Calculate difference in review scores and square them 
                sim_score = sim_score + sim_score_temp           # Increment the score counter 
        sim_score = sim_score ** 0.5                             # Square root of score counter 
        print('The similarity score between {} and {} is: {}'.format(keys_temp1, keys_temp2, round(sim_score,4)))
        print(' ')                                               # Print new line   
    return()                                                     # Print the similarity score and return back  

# Function when option 2 is selected 
def option_2(f_dict):                                            # Accept the dictionary parameter 
    rev = input('Provide Reviewer name: ')                       # Prompt for reviewer name   
    if rev.isspace() or len(rev) == 0:                           # Check for invalid entry in reviewer name 
        print('Reviewer name not entered') 
    elif not (case_insensitive(f_dict, rev)):                    # Check if reviewer name is present in database 
        print('Reviewer is not present in database') 
    else:                                                        # If reviewer is present in database, then execute below 
        dict_rev , keys_temp1 = build_subdict(f_dict , rev)      # Retrieve the movies and thier scores in separate dict -
        print('The Similarity Scores are: ')                     # - by calling the build subdict function 
        for keys, values in f_dict.items():                      # Loop thru the main dict
            sim_score_temp = 0                                   # Initialize score counters
            sim_score = 0                                          
            if rev.lower() == keys.lower():                      # Skip comparison for entered reviewer name with themselves    
                continue 
            else: 
                for keys1, values1 in values.items():            # Loop thru movies names & scores for specific reviewer  
                    if keys1 in dict_rev:                        # For matching movie names, calculate similarity score
                        sim_score_temp = (dict_rev[keys1] - values1) ** 2 # Take score difference and square them
                        sim_score = sim_score + sim_score_temp   # Increment the score counter 
                sim_score = sim_score ** 0.5                     # Similarity score between 2 reviewers. Taking square root 
                print('%-20s %-20s %4.2f'%(keys_temp1, keys, sim_score)) # Display similarity score between 2 reviewers
                #print('{}     {}          {}'.format(keys_temp1, keys, round(sim_score,2)))
    print(' ')                                                   # Print new line
    return()                                                     # End function execution and return    

# Main program -> Body of Program 
file_name = input('Give the name of the movie reviews file: ')   # Prompt user to enter input file name 

try: 
    file_handle = open(file_name,'r')                            # open the entered input file in read mode  
except:                                                          # Handle exception when file name is incorrect or -  
    print('Invalid file name / not present in directory')        # - present in invalid directory location 
else:   
    dictionary , comma_count = build_dict(file_handle)           # Read thru file and build a dictionary
    if comma_count < 0:                                          # Invalid file format cases
        if comma_count == -1:                                    # When comma delimiter is missing or present only once   
            print('Invalid format of records in Input file. Missing comma delimiter')
        else:                                                    # When movie review score is not numeric value    
            print('Invalid format of records in Input file for reviewer movie score') 
    elif comma_count == 0:                                       # When no records are present in input file 
        print('Empty input file. Please check')
    else:                                                        # When input file records are in correct format  
        while True:                                              # Loop thru until user chooses to exit 
            user_choice = prompt_choice()                        # Prompt user for input choice
            if user_choice == '1':                               # When user chooses Option 1 
                option_1(dictionary)                             # Call option 1 function by passing dictionary 
            elif user_choice == '2':                             # When user chooses Option 2 
                option_2(dictionary)                             # Call option 2 function by passing dictionary 
            else:                                                # When user chooses Option 3    
                print('Goodbye!')                                
                break                                            # Break from the loop and end program execution 

# End of Program 2
```
