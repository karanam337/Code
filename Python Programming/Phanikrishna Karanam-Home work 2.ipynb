#### __Question 1__:  
__Summary__: A program that rolls a die 100 times and displays the number of 1s, 2s, 3s, 4s, 5s and 6s. It also provide a graphical display of the frequencies.

##### __Execution Instructions__:
1. Run the cell below   
2. To rerun the test, execute the code cell again 

##### __Expected Results__:
 - Display the summary stats on which number rolled how many times, in both number and graphical format

##### __Algorithm__:
` 1. Generate 100 random integers from 1 to 6, to mimick a die roll 100 times and store them into a list `  
` 2. Count the number of occurances each number has appeared within the list using count function and display the stats `  
` 3. Multiply * using count number of times, to display the results graphically `  

```python
# Program for question 1

import random                                   # import the random function

die_roll = list()                               # list to store the results of die roll 
for i in range(100):                            # generate 100 random integers from 1 to 6 and store into list
    die_roll.append(random.randint(1,6))

print("Frequencies: ")
counter = 0
for i in range(1,7):                            # determine count of each die roll from 1 to 6 and display the results 
    counter = die_roll.count(i)
    print("Number of {}s:".format(i),counter)
    
for i in range(1,7):                            # determine count of each die roll from 1 to 6, display graphical results  
    counter = die_roll.count(i)
    print("{}     ".format(i),('*' * counter))

# End of program for question 1
```

#### __Question 2__:  
__Summary__: The program prompts user for a number (between 1 and 100, inclusive) and displays the number of iterations it took for the program to guess the number. 

##### __Execution Instructions__:
1. Run the cell below   
2. Enter a number from 1 to 100
3. If invalid input entered, execute the code cell again and follow the prompt
4. To rerun the test, execute the code cell again and follow the prompt

##### __Expected Results__:
 - Display the user entered number and number of iterations the program took to guess the number 
 - For invalid inputs, error message is displayed

##### __Algorithm__:
` 1. Prompt user for an input number from 1 to 100 and convert the input to integer `  
` 2. For invalid inputs:
     2.1. If non numeric entry, display error message
     2.2. If entered number is outside of desired range of 1 to 100, then display error message `  
` 2. For valid input:
     3.1. Generate a random number from 1 to 100
     3.2. Increment the guess counter by 1
     3.3. If input number is same as random number generated, then break the loop. Else repeat steps from 3.1 until a match `   
` 4. Display the entered number and number of iterations taken to guess the input number `  

```python
# Program for question 2

import random                                                                 # Import the random function  

try: 
    user_input = int(input("Enter a number from 1 to 100, to be guessed: "))  # Accept an input number from user
except ValueError: 
    print("Invalid input. Enter valid integer from 1 to 100")                 # For non numeric entries, display erorr
else: 
    if ((user_input >= 1) and (user_input <= 100)):                           # Check if input number is from 1 to 100
        counter = 0                                                           # Check to avoid infinite loop possibility
        while True:                                                           # Loop to generate random integers 1 to 100
            computer_guess = random.randint(1,100)                             
            counter = counter + 1                                             # Counter to track the number of iterations
            if user_input == computer_guess:                                  # If input number matches to random number,  
                break                                                         # then break from loop & display message  
        print("You entered {}, and it took program {} iterations to guess".format(user_input,counter))
    else:
        print("Invalid input. Enter valid integer from 1 to 100")             # If input number is outside of 1 and 100 

# End of program for question 2
```

#### __Question 3__:  
__Summary__: A function that accepts a line of text and a single letter as input and returns the number of times the letter is the first character of a word

##### __Execution Instructions__:
1. Run the cell below   
2. Enter text (a sentence) at first prompt and single letter at second prompt 
3. If invalid (blank or space) input entered at first prompt, then execute the code cell again and follow the prompts
5. If invalid (blank, space or more than one character) input entered at second prompt, then execute the code cell again and follow the prompts
6. To rerun the test, execute the code cell again and follow the prompts

Note: Inputs are case sensitive

##### __Expected Results__:
 - Display how many times the input letter (entered on second prompt) is the first character of a word from the text (entered on first prompt) 

##### __Algorithm__:
` 1. Define a function that will accept 2 parameters as inputs
     1.1. First parameter is a List that will contain text entered (first input), with each word as separate list item
     1.2. Second parameter is the single character letter entered by user 
     1.3. Count how many times the input letter is present as first letter of each word in input text 
     1.4. Return value of this count as function output `  
` 2. Accept the input line of text from user
     2.1. Split the entered text into a list of words using split function
     2.2. If blank line or spaces entered, then display error message 
     2.3. If valid line of text entered:  
          2.3.1. Accept a letter from user. If invalid entry (like space or more than one character entered), then display
                 error message
          2.3.2. If valid input (single non blank letter) received from user, call function defined in step 1 by passing the
                 list containing all words entered as part of first input and letter entered in second input as arguments
          2.3.3. Display the output message using the return value from the function call `  

```python
# Program for question 3

# Below function will accept 2 parameters as input values 
# First parameter contains a list having all the words that were entered by user on first input
# Second parameter contains a letter entered by user on second input
# This function will count how many times the letter entered (on second input) is present as first character of every - 
#  - word entered on the first input. This count will be returned as the function output

def word_count(line, first_letter):
    count = 0
    for word in line: 
        if (first_letter == word[:1]):
            count = count + 1                                 # counter tracking number of words with matching first letter
    return(count)

# Below function is the main body of the program which will call the word_function 
# Accepts two inputs from user
# It displays how many times a letter (second input) is present as first word in the text entered (first input)

def main():
    line_inp = input("Enter your line of text: ")             # Accept first input from user 
    line_list = line_inp.split()                              # Convert the string to a list to separate all the words
    if (len(line_list) > 0):                                  # Check if input string is valid
        letter_inp = input("Enter your letter to use: ")      # Prompt user to enter the letter as second input
        if ((len(letter_inp) == 1) and (letter_inp != ' ')):  # Check if entered input is one character and not a space
            first_lcount = word_count(line_list, letter_inp)  # Call word_count function by passing the necessary arguments
            if (first_lcount > 1):                            # If return count is plural, display as times v/s time
                print("Your letter '{}' occurs as the first letter: {} times".format(letter_inp, first_lcount))
            else: 
                print("Your letter '{}' occurs as the first letter: {} time".format(letter_inp, first_lcount))
        else:                                                 # Error message on second input error
            print("More than one / invalid letter to use entered. Enter single letter only") 
    else:                                                     # Error message on first input error
        print("No text entered. Enter valid line of text")     

main()                                                        # Invoke the body of program

# End of program for question 3
```

##### __Expected Results__:
 - Strings are balanced only if all characters from string1 are present on string2 (irrespective of the position) 
     - Weight of the characters (number of occurances) in string1 should be equal or less than that of string2
 - Strings are NOT balanced if all characters from string1 are not present on string2 (irrespective of the position) 
     - Weight of the characters (number of occurances) in string1 is greater than that of string2
 - When blank or spaces are entered as inputs, then error messages will be displayed 
     
Note: Vice-versa (string2 vs string1) is not being considered 
         
##### __Algorithm__:
` 1. Define a function that will accept 1 parameter as input
     1.1. Entered string is passed into the function and used to build a dictionary 
     1.2. Store every letter of the string as an item along with its occurance value in that string 
     1.3. Return the dictionary containing every letter of the string and thier occurance count as output of function `  
` 2. Accept first input string1 from user
     2.2. If blank line or spaces entered, then display error message 
     2.3. If valid line of text entered:  
          2.3.1. Accept second input string2 from user. If invalid entry (blank line or space), then display error message
          2.3.2. If valid input received from user, call function defined in step 1 twice by passing the input strings as
                 argument each time. Two dictionaries will be built based on input string characters and their weights
          2.3.3. For every character in string1, If:
                 2.3.3.1. NOT (If its present in string2 and weight of each character is atleast <= to
                 weight of characters in string 2), then the strings are NOT balanced. Break from the loop on first occurance
                 2.3.3.2. Else the strings are balanced `   

```python
# Program for question 4

# Below function will accept a string as parameter and build a dictionary using the same
# Every character in the string will be stored as an item in the dictionary along with the number of times the character - 
# - has appeared in the string. The function will return this dictionary as output, which will contain every character - 
# - and their weights (or number of occurances)

def build_dict(inp):
    dictionary = {}
    for i in inp: 
        if i in dictionary: 
            dictionary[i] = dictionary[i] + 1
        else: 
            dictionary[i] = 1
    return(dictionary)

def main():
    inp1 = input('Enter String1: ')                                           # Prompt for string1  
    if (len(inp1) > 0 and not(inp1.isspace())):                               # If string1 is not blank or spaces   
        inp2 = input('Enter String2: ')                                       # Prompt for string2  
        if (len(inp2) > 0 and not(inp2.isspace())):                           # If string2 is not blank or spaces     
            inp_dict1 = build_dict(inp1)                                      # Build dictionary by calling function and -  
            inp_dict2 = build_dict(inp2)                                      # - passing input string as argument  
            for i in inp_dict1:                                               # loop thru for every character in string1 
                if not ((i in inp_dict2) and (inp_dict1[i] <= inp_dict2[i])): # Refer below to why negative condition test?
                    print("String1 and String2 are NOT balanced")
                    break                                                     # Break from for loop if condition is met 
            else:
                print("String1 and String2 are balanced")                     # Will execute if for loop is not terminated  
        else:
            print("Invalid String2 entered")                                  # error message for invalid string2 input
    else:
        print("Invalid String1 entered")                                      # error message for invalid string1 input  


main()                                                                        # Invoke body of the program   

# Why negative condition test?
# Positive condition needs to be checked for every character of string1, but the negate of this would directly satisfy - 
# - the NOT balanced check even if met for one character of string1 and thus concluding strings are not balanced. 
# If negative condition is not met for all the characters, that means loop is not terminated by break condition and - 
# - strings are balanced, which is in the else part of for loop

# End of program for question 4
```

#### __Question 5__:  
__Summary__: A program to request a file name from the user and calculate below statistics of the contents of the file:
    1. Number of lines (is a sequence of characters that end with a newline (\n) character, except for last line)
    2. Number of words (bounded by one or more spaces (or \n) on either side of it or both sides)
    3. Number of characters (any single length string, e.g. ‘a’, ‘-‘, etc. but not a space or white space)
    4. Average length of a word

##### __Execution Instructions__:
1. Run the cell below   
2. Enter the file name along with extension (Example: RobertFrost.txt) 
3. Ensure the input file is present in same directory where the python code is being executed from
4. To rerun the test, execute the code cell again and follow the prompt

##### __Expected Results__:
 - Displays the number of lines, words, characters and average length of the word in the input file
     - Empty lines, blanks or spaces in the file are not counted as lines 
     - Space and new line character are not counted as a character. Except them all others (like comma, fullstop) are counted 
     - Average length of the word is rounded upto 1 decimal point
 - If file is empty, then error message is displayed stating empty file
 - If incorrect file name is entered or if file is not present in directory (same as where code is being executed from), 
   then error message will be displayed 
         
##### __Algorithm__:
` 1. Prompt user for file name `  
` 2. Open the file in read mode 
     2.1. If any exceptions (like incorrect file name or file not present in directory), then display error message `  
` 3. Initialize all the counters to be used in the program `  
` 4. Loop thru the file, line wise using a for loop until end of file. This will automatically take care of reading until 
     new line character (marking end of line) except for the last line (which is end of file)
     4.1. If the line is blank or spaces, then skip that line and proceed to next line
     4.2. If line is not blank then: 
          4.2.1. Increment the line counter
          4.2.2. Convert the input line (string) to individual words (list) using split function. This will automatically 
                 remove the spaces within each words and also the new line character at the end of every line
          4.2.3. Calculate the word count by calculating the length of the list. Increment and keep track for each line
          4.2.4. Loop thru the list to calculate the length of each word. Increment the counter by adding length for every 
                 word and keep track of the number of characters `   
` 5. If word counter is greater than zero (indicating file is not empty): 
     5.1. Calculate the average word length by dividing character count by word count. Round upto 1 decimal point
     5.2. Display the input file statistics -> line, word, character counts and average length of a word `  
` 6. If word counter is not greater than zero, then display empty file error message `   
` 7. Close the input file `   

```python
# Program for question 5

file_name = input('What is the file name: ')                      # Prompt user to enter input file name 

try: 
    file_handle = open(file_name,'r')                             # open the entered input file in read mode 
    
    line_count = character_count = word_count = average_word = 0  # Initialize all counters

    for line in file_handle:                                      # loop thru the file. Reads one line (upto \n) at a time 
        if line.isspace():                                        # If the line is empty or spaces, then skip processing 
            continue 
        else:                                                     
            line_count = line_count + 1                           # increment line counter for each non blank line
            line_temp = line.split()                              # split each line (string) into words (list)
                                                                  # Spaces and /n characters are automatically removed 
            word_count = word_count + len(line_temp)              # Track the count of number of words for each line
            for word in line_temp:                                # Loop thru each word to track count of each character -  
                character_count = character_count + len(word)     # - on a line read
             
    if word_count > 0:                                            # Check to avoid zeroDivisionError when input file empty
        average_word = round(character_count/word_count,1)        # Calculate the average word length in the file
        print('Number of lines:', line_count)                     # Display required file statistics
        print('Number of words:', word_count)
        print('Number of characters:', character_count)
        print('Average length of a word:', average_word) 
    else: 
        print('Input file is empty')                              # Error message when inpyt file is empty 

    file_handle.close()                                           # Close the file  
except:                                                           # Handle exception when file name is incorrect or -  
    print('Invalid file name / not present in directory')         # - present in invalid directory location 

# End of program for question 5
```
